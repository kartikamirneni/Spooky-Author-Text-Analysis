```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
import string

%matplotlib inline

eng_stopwords = set(stopwords.words("english"))
pd.options.mode.chained_assignment = None
```

```python
## Read the train and test dataset and check the top few lines ##
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
print("Number of rows & columns in train dataset : ",train_df.shape)
print("Number of rows $ columns in test dataset : ",test_df.shape)
```

## 3. Average length of the words in the text ##

```python
print("Average length of words in the training dataset")
train_df["mean_word_len"] = train_df["text"].apply(lambda x: np.mean([len(i) for i in str(x).split()]))
print(train_df["mean_word_len"])

train_bins=20
arr=plt.hist(train_df["mean_word_len"],bins=train_bins)
for i in range(train_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

```python
print("Average length of words in the testing dataset")
test_df["mean_word_len"] = test_df["text"].apply(lambda x: np.mean([len(i) for i in str(x).split()]))
print(test_df["mean_word_len"])

test_bins=20
arr=plt.hist(test_df["mean_word_len"],bins=test_bins)
for i in range(test_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

## 4. Number of words in the text ##

```python
print("Number of words in the training dataset")
train_df["num_words"] = train_df["text"].apply(lambda x: len(str(x).split()))
print(train_df["num_words"])

train_bins=20
arr=plt.hist(train_df["num_words"],bins=train_bins)
for i in range(train_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

```python
print("Number of words in the testing dataset")
test_df["num_words"] = test_df["text"].apply(lambda x: len(str(x).split()))
print(test_df["num_words"])

test_bins=20
arr=plt.hist(test_df["num_words"],bins=test_bins)
for i in range(test_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

## 5. Number of unique words in the text ##

```python
print("Number of unique words in the training dataset")
train_df["unique_words"] = train_df["text"].apply(lambda x: len(pd.unique(str(x).split())))
print(train_df["unique_words"])

train_bins=20
arr=plt.hist(train_df["unique_words"],bins=train_bins)
for i in range(train_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

```python
print("Number of unique words in the testing dataset")
test_df["unique_words"] = test_df["text"].apply(lambda x: len(pd.unique(str(x).split())))
print(test_df["unique_words"])

test_bins=20
arr=plt.hist(test_df["unique_words"],bins=test_bins)
for i in range(test_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

## 6. Number of stop words in the text ##

```python
print("Number of stop words in the training dataset")
train_df["stop_words"] = train_df["text"].apply(lambda x: len([i for i in str(x).lower().split() if i in eng_stopwords]))
print(train_df["stop_words"])

train_bins=20
arr=plt.hist(train_df["stop_words"],bins=train_bins)
for i in range(train_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```

```python
print("Number of stop words in the testing dataset")
test_df["stop_words"] = test_df["text"].apply(lambda x: len([i for i in str(x).lower().split() if i in eng_stopwords]))
print(test_df["stop_words"])

test_bins=20
arr=plt.hist(test_df["stop_words"],bins=test_bins)
for i in range(test_bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
```
